```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
```

```python
pd.set_option('display.max_columns',79)
df = pd.read_csv('data\survey_results_public.csv')
df

```

```python
RemoteWork = df['RemoteWork']
```

```python
dataRemoteWork = dict(RemoteWork.value_counts())
```

```python
WorkType = list(dataRemoteWork.keys())
counting = list(dataRemoteWork.values())
```

```python
colors =['#ffd700','#87cefb','#9acd32']
plt.pie(counting,labels=WorkType,shadow=True,colors=colors,autopct='%.2f%%',wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
LanguageWantToWorkWith= df['LanguageWantToWorkWith']
language_counter = Counter()
```

```python
LanguageWantToWorkWith =dict(LanguageWantToWorkWith)
```

```python
language =list(LanguageWantToWorkWith.values())
```

```python
for x in language:
    x = str(x)
    if x =="nan":
        continue
    language_counter.update(x.split(";"))
```

```python
lan = []
propularity =[]
for item in language_counter.most_common(20):
    lan.append(item[0])
    propularity.append(item[1])
    
```

```python
lan.reverse()
propularity.reverse()
plt.barh(lan,propularity,height=0.5)
plt.tight_layout()
plt.show()
```
